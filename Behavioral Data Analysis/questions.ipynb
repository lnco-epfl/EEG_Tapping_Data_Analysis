```python
import json

# Load the JSON file
with open("data_P01.json", "r", encoding="utf-8") as file:
    data = json.load(file)

print(data)
# Define the questions to track
target_questions = {
    "I felt I was in control of the bar's movement.": [],
    "I found the task difficult.": []
}

# Extract relevant responses in order
for entry in data:
    if entry.get("trial_type") == "survey-likert":
        for question in target_questions:
            if question in entry.get("response", {}):
                target_questions[question].append(entry["response"][question])

# Print or save the extracted responses
print(json.dumps(target_questions, indent=2, ensure_ascii=False))
```

```python
# Define the list of indices without delay
ND = [1, 2, 5, 7]

# Define the list of indices with delay
D = [0, 3, 4, 6]

# Q1
print("Q1: I felt I was in control of the bar's movement.")
print("Responses without delay:")
for idx in ND:
    print(target_questions["I felt I was in control of the bar's movement."][idx])

print("Responses with delay:")
for idx in D:
    print(target_questions["I felt I was in control of the bar's movement."][idx])


# Q2
print("Q2: I found the task difficult.")
print("Responses without delay:")
for idx in ND:
    print(target_questions["I found the task difficult."][idx])

print("Responses with delay:")
for idx in D:
    print(target_questions["I found the task difficult."][idx])
```
